#### Code the solution

```python
data = pd.read_csv('homework3_input_data.csv')
data
```

```python
x_train, x_test, y_train, y_test = train_test_split(data[['cement','slag','flyash','water','superplasticizer','coarseaggregate','fineaggregate','age']], data['csMPa'], test_size=0.2, random_state=0)
```

```python
data.describe()
```

```python
pyplot.figure(figsize=(30,20))
for i, col in enumerate(data.columns[0:8]):                                                
    pyplot.subplot(5, 3, i+1) 
    x = data[col] 
    y = data['csMPa']
    pyplot.plot(x, y, '.', color="forestgreen")
    pyplot.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)),color="red")
    pyplot.xlabel(col)
    pyplot.ylabel('Cement')
```

```python
features = data[['cement','slag','flyash','water','superplasticizer','coarseaggregate','fineaggregate','age']]
sns.set(rc={'figure.figsize': (8.5,8.5)})
sns.heatmap(features.corr().round(2), square=True, cmap='YlGnBu', annot=True)
```

```python
train_df = pd.DataFrame(x_train,columns = ['cement','slag','flyash','water','superplasticizer','coarseaggregate','fineaggregate','age'])
train_df['csMPa'] = y_train
sns.pairplot(train_df, vars = ['cement','slag','flyash','water','superplasticizer','coarseaggregate','fineaggregate','age'])
```

```python
model = linear_model.LinearRegression().fit(x_train, y_train)
```

```python
print('Coefficients:')
pp(model.coef_)
y_test_pred = model.predict(x_test)
print(f'Mean Squared Error: {mean_squared_error(y_test, y_test_pred)}')
print(f'Coefficient of Determination: {r2_score(y_test, y_test_pred)}')
```
